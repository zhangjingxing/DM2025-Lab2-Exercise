## 1. Model Development (10 pts Required)
    

### 1.1 Preprocessing Steps

The text preprocessing pipeline transforms the raw posts into a clean, model-ready format. The process begins by standardizing the text through lowercasing and expanding common contractions like "don't" to "do not". Crucially for emotion analysis, it converts emojis into descriptive text (e.g., '\_heart\_') rather than removing them. It then cleans the data by stripping out noise like URLs, @mentions, and punctuation. Finally, the text is tokenized, and a custom list of stop words is removed, while purposefully preserving negation words (like "not"), before each word is lemmatized to its dictionary form. The cleaned tokens are then re-joined into a single string, ready for feature extraction.

### 1.2 Feature Engineering Steps

For the feature engineering step, it needed to convert our cleaned text data into a numerical format that machine learning models can understand. I chose the TF-IDF (Term Frequency-Inverse Document Frequency) vectorization method. This approach is highly effective for text classification because it creates a matrix that scores words based not just on their frequency in a single post, but also on their rarity across the entire dataset, effectively highlighting important, predictive keywords. We specifically set ngram_range=(1, 2) to capture both individual words (unigrams) and two-word pairs (bigrams), allowing the model to learn crucial phrases like "not happy" as a single feature. We also limited the feature set to the top max_features=5000 to save memory and focus the model on the most relevant terms.

Besides, cause later i will try CNN stuff, I also try different embeddings, not from scratch and instead used GloVe-6B. This is a large set of pre-trained word vectors trained on 6 billion words, which gives the model a built-in "head start" on understanding language (like "happy" is similar to "joy"). 

### 1.3 Explanation of Your Model

My final and most successful model, which achieved the 0.6595 score, was a DistilBERT Transformer. This represents a completely different and far more powerful approach to the problem. Instead of relying on hand-crafted features or "bag-of-words" models, DistilBERT is a state-of-the-art model that is pre-trained on a massive amount of text. Its key advantage is that it understands context. It doesn't just see the word "bank"; it knows whether that word means a river bank or a money bank based on the other words in the sentence.

The preparation for this model was unique. Unlike all my previous efforts where I carefully cleaned and pre-processed the text, the DistilBERT model performed best on the original, raw text. This is because it uses signals I had previously thrown away, such as punctuation (!), capitalization (ALL CAPS), and stop words ("to", "the", "a"), as part of its analysis. I used the specific DistilBertTokenizer from the Hugging Face transformers library. This tokenizer converts the raw text into "Input IDs" (numerical sequences) and, just as importantly, creates an "Attention Mask" that tells the model which tokens are real words versus padding.

I loaded the distilbert-base-uncased model, which already possessed a deep knowledge of the English language. I then attached a new classification "head" to it, with six outputs for my six emotion classes. The critical part of this step was compiling the model. It required a specific optimizer, AdamWeightDecay, and a very low learning rate (like 5e-5). This slow-and-steady approach allowed the model to adapt its massive pre-trained knowledge to my specific task of emotion classification without "forgetting" its original understanding of language.

This fine-tuned Transformer model was the clear winner because it was the only one that could truly read the text for its nuanced meaning. While my TF-IDF models just counted important words and my CNNs learned word patterns, the DistilBERT model understood the full contextual meaning of the posts. This deep understanding allowed it to capture the subtle cues of emotion far more effectively than any other method, leading to the breakthrough in my score.

## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section
import json
import pandas as pd

# load the dataset
json_file_path = 'D:/Disk E/Jeremy/NTHU/Kuliah/Sem 7/Data Mining/HW/DM2025Labs/dm-lab-2-private-competition/final_posts.json'
csv_splits_path = 'D:/Disk E/Jeremy/NTHU/Kuliah/Sem 7/Data Mining/HW/DM2025Labs/dm-lab-2-private-competition/data_identification.csv'
csv_labels_path = 'D:/Disk E/Jeremy/NTHU/Kuliah/Sem 7/Data Mining/HW/DM2025Labs/dm-lab-2-private-competition/emotion.csv'

post_ids = []
texts = []

with open(json_file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

for entry in data:
    try:
        post_data = entry['root']['_source']['post']
        post_ids.append(post_data['post_id'])
        texts.append(post_data['text'])
    except KeyError as e:
        print(f"Skipping an entry due to missing key: {e}")


df_text = pd.DataFrame({'id': post_ids, 'text': texts})

df_splits = pd.read_csv(csv_splits_path)
df_labels = pd.read_csv(csv_labels_path)

# merge dataframes
df_merged = pd.merge(df_text, df_splits, on='id', how='inner')
df_all_data = pd.merge(df_merged, df_labels, on='id', how='left')

# separate train and test set
train_df = df_all_data[df_all_data['split'] == 'train'].dropna(subset=['emotion']).copy()
test_df = df_all_data[df_all_data['split'] == 'test'].copy()

train_df = train_df.drop(columns=['split'])
test_df = test_df.drop(columns=['split'])

print(train_df)
print(test_df)

```

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import emoji

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
negation_words = {'no', 'not', 'never', 'don', "don't", 'ain', 'aren', "aren't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'mightn', "mightn't", 'mustn', "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', "won't", 'wouldn', "wouldn't"}

final_stop_words = stop_words - negation_words

lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    # 1. Lowercasing
    text = text.lower()
    
    # 2. Expand contractions (a simple example dictionary)
    contractions = { 
        "i'm": "i am", "you're": "you are", "he's": "he is", "she's": "she is",
        "it's": "it is", "we're": "we are", "they're": "they are", "don't": "do not",
        "can't": "cannot", "won't": "will not", "shouldn't": "should not"
        # Add more contractions as needed
    }
    for k, v in contractions.items():
        text = text.replace(k, v)
    
    # 3. Convert emojis to text
    text = emoji.demojize(text, delimiters=("_", "_")) 

    # 4. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # 5. Remove @mentions
    text = re.sub(r'@\w+', '', text)
    
    # 6. Remove punctuation
    # We keep underscores so the emoji text isn't split
    punctuation_to_remove = string.punctuation.replace("_", "")
    text = text.translate(str.maketrans('', '', punctuation_to_remove))
    
    # 7. Tokenization
    tokens = word_tokenize(text)
    
    # 8. Remove stop words and lemmatize
    cleaned_tokens = []
    for token in tokens:
        if token not in final_stop_words:
            # Lemmatize and add to the list
            cleaned_tokens.append(lemmatizer.lemmatize(token))
            
    # 9. Join tokens back into a single string
    return " ".join(cleaned_tokens)
```

```python
train_df['clean_text'] = train_df['text'].apply(preprocess_text)
test_df['clean_text'] = test_df['text'].apply(preprocess_text)
```

### Embedding for our CNN

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

VOCAB_SIZE = 10000    # Max words to use
MAX_LEN = 150         # Max length to pad/truncate sentences to
EMBEDDING_DIM = 100   

# y is still train_df['emotion']
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
num_classes = len(label_encoder.classes_)

X_text = train_df['clean_text']
X_test_text = test_df['clean_text']

tokenizer = Tokenizer(num_words=VOCAB_SIZE, oov_token="<OOV>")
tokenizer.fit_on_texts(X_text)


X_sequences = tokenizer.texts_to_sequences(X_text)
X_test_sequences = tokenizer.texts_to_sequences(X_test_text)

X_padded = pad_sequences(X_sequences, maxlen=MAX_LEN, padding='post', truncating='post')
X_test_padded = pad_sequences(X_test_sequences, maxlen=MAX_LEN, padding='post', truncating='post')

X_tr, X_val, y_tr, y_val = train_test_split(
    X_padded,
    y_encoded,
    test_size=0.2,
    random_state=42
)

print(f"Padded training data shape: {X_tr.shape}")
print(f"Padded validation data shape: {X_val.shape}")
print("Data preparation complete.")
```

```python
embeddings_index = {}
glove_file_path = 'glove.6B.100d.txt' 

try:
    with open(glove_file_path, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = coefs
    print(f"Found {len(embeddings_index)} word vectors.")
except FileNotFoundError:
    print("ERROR: glove.6B.100d.txt not found.")
    
```

```python
word_index = tokenizer.word_index
num_words = min(VOCAB_SIZE, len(word_index) + 1)
embedding_matrix = np.zeros((num_words, EMBEDDING_DIM))

for word, i in word_index.items():
    if i >= VOCAB_SIZE:
        continue
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector

print("Embedding matrix created.")
```

### This is for our next experiment, TRANSFORMER

```python
import tensorflow as tf
from transformers import DistilBertTokenizer, TFDistilBertForSequenceClassification
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd

X_text = train_df['text']
X_test_text = test_df['text']

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(train_df['emotion'])
num_classes = len(label_encoder.classes_)

# We split the text itself, not the arrays
X_tr_text, X_val_text, y_tr, y_val = train_test_split(
    X_text,
    y_encoded,
    test_size=0.2,
    random_state=42
)

tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
print("Tokenizing data...")
X_tr_encodings = tokenizer(
    list(X_tr_text), truncation=True, padding=True, max_length=128
)
X_val_encodings = tokenizer(
    list(X_val_text), truncation=True, padding=True, max_length=128
)
X_test_encodings = tokenizer(
    list(X_test_text), truncation=True, padding=True, max_length=128
)

# This is the format the model wants for training
train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(X_tr_encodings),
    y_tr
)).batch(16)

val_dataset = tf.data.Dataset.from_tensor_slices((
    dict(X_val_encodings),
    y_val
)).batch(16)

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(X_test_encodings)
)).batch(16)

print("Data preparation complete.")
```

```python
import tensorflow as tf
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from transformers import TFDistilBertForSequenceClassification
from transformers import AdamWeightDecay  

# Load the model
model_bert = TFDistilBertForSequenceClassification.from_pretrained(
    'distilbert-base-uncased', 
    num_labels=num_classes, 
    use_safetensors=False
)

# Use the correct optimizer
optimizer = AdamWeightDecay(learning_rate=5e-5) 
loss = SparseCategoricalCrossentropy(from_logits=True)
model_bert.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])

# Train the model
print("Starting BERT fine-tuning...")
history_bert = model_bert.fit(
    train_dataset, 
    epochs=3,
    validation_data=val_dataset 
)

print("Model training finished!")
```

```python
predictions_logits = model_bert.predict(test_dataset).logits

final_predictions_num = np.argmax(predictions_logits, axis=1)

final_predictions_labels = label_encoder.inverse_transform(final_predictions_num)

submission_bert_df = pd.DataFrame({
    'id': test_df['id'],
    'emotion': final_predictions_labels
})

submission_bert_df.to_csv('Submission_BERT.csv', index=False)
```
